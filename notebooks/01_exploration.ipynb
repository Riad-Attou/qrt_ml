```python

# Selecting only numeric columns from the training dataset
numeric_cols = clinicaltrain.select_dtypes(include=['number']).columns

# Get correlation of all the numeric features of the training dataset
corr_matrix = clinicaltrain[numeric_cols].corr()
top_corr_features = corr_matrix.index

# Plotting the heatmap
plt.figure(figsize=(20,20))
sns.heatmap(data=clinicaltrain[top_corr_features].corr(), annot=True, cmap='RdYlGn')
plt.show()
```
